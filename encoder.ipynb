# Encoding gene seq in Brainfuck syntax

"A" --> 1 or {'+' in brainfuck}
"C" --> 2 or {'++' in brainfuck}
"G" --> 3 or {'+++'}
"T" --> 4 or {++++}


```python
# written on 26/07/17

import sys
import string
```

```python
gene = open("BRCA1_BAP1_copy.txt","r")
```

```python
st = gene.readline()
```

```python
for line in gene:
    line = line.lower()
    for char in line:
        if char == "a":
            st = st + '+>'
        if char == "c":
            st = st + '++>'
        if char == "g":
            st = st + '+++>'
        if char == "t":
            st = st + '++++>'
            
```

```python
print(st)
```

```python
text_file = open("Output.txt", "w")
text_file.write(st)

text_file.close()
```
